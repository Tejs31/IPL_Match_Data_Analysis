```python
#Distribution plot - We use distribution plot to visulize the distribution of quantitative data.
#A distribution plot (distplot) shows how the values of a numerical variable are distributed.

#It helps you understand:

#Where most values lie
#Whether data is normally distributed
#Skewness (left/right skew)
#Outliers
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
sns.displot(df['total_bill'],bins=20,kde=True) #or sns.histplotplot(df['total_bill'],bins=20)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
#jointplot() is used to visualize the relationship between two numerical variables and their distributions in a single figure.
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
#pairplot() is used to visualize relationships between multiple numerical variables in a dataset.
sns.pairplot(df,hue='sex')
```

```python
#Categorical plot: catplot() (Categorical Plot) is used to visualize relationships between categorical and numerical variables.
#A count plot is used to visualize the number of occurrences of each category in a categorical variable. It displays category frequencies using bars.
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
#A Bar Plot is used to show the relationship between a categorical variable and a numerical variable.
#The height of each bar represents an aggregated value (by default, the mean).
sns.barplot(x=df['sex'],y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'],y=df['tip'],estimator=np.sum)
```

```python
#A box plot displays the distribution of data using the five-number summary: minimum, Q1, median, Q3, and maximum, and helps identify outliers.
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

```python
#A Strip Plot is used to visualize the distribution of a numerical variable across different categories.
#Unlike a bar plot or box plot, it shows every individual data point.
#A Swarm Plot is similar to a Strip Plot, but it automatically adjusts points so they do not overlap.
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
#In Seaborn, a Matrix Plot is used to visualize data in a matrix format, usually to identify patterns, correlations, or missing values.
#HeatMap:Used to visualize correlations or values in a matrix.
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipcorr=tips[['total_bill','tip','size']]
tipcorr
```

```python
tipcorr.corr()
```

```python
sns.heatmap(tipcorr.corr(),annot=True)
```

```python
sns.clustermap(tipcorr.corr())
```

```python
#A Pivot Table Heatmap combines:
#Pivot Table → Summarizes data.
#Heatmap → Uses colors to visualize the summarized values.
#It helps quickly identify patterns, high values, and low values.
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```

```python
#Regression plot: A Regression Plot is used to visualize the relationship between two numerical variables along with a best-fit regression line.
#It helps determine:
#Whether variables are positively correlated
#Negatively correlated
#Or have no relationship
#lmplot() stands for Linear Model Plot.
#It is used to visualize the relationship between two numerical variables and draw a regression line.
tips=sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
df=pd.read_csv('IPL.csv')
df
```

```python
#Baic Information
df.info()
```

```python
#Check the size and rows of column 
df.shape
```

```python
#How many column have null values in total 
df.isnull().sum()
```

```python
#1.Which team won the most matches?
match_wins=df['match_winner'].value_counts()
match_wins
```

```python
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title('Most match win by each team')
```

```python
#2.Toss Decision Trend
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title('Toss Decision Trend')
```

```python
#3.Toss Winner vs Match Winner
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

```python
#4.How do teams wins(Runs or wickets)
sns.countplot(x=df['won_by'])
plt.title('Won by')
```

```python
#5.Most Player of the match
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y=count.index,palette='mako')
plt.title('Player of the Match')
```

```python
#6.Top 2 Scorer 
high=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
top=high.head(2)
top
```

```python
top.plot(kind='barh')
```

```python
#7.10 Best Bowling figure 
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split("--")[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='bar')
```

```python
#8. Most matches played at Venue
venue_count=df["venue"].value_counts()
venue_count
```

```python
sns.barplot(y=venue_count.index,x=venue_count.values,palette='rainbow')
```

```python
#9.Who won the highest margin by run 
df[df['won_by']=='Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','margin']]
```

```python
#10.Which Player had the highest indivisual score 
df[df['highscore']==df['highscore'].max()][['top_scorer','highscore']]
```

```python
#11.Which bowler has the best bowling figure
df[df['highest_wickets']==df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```
